```python
# Set the frequency of the date time index as Monthly start as indicated by the data
data.index.freq = 'MS'
# Set the value of Alpha and define m (Time Period)
m = 15
alpha = 1/(2*m)
```

```python
data['HWES1'] = SimpleExpSmoothing(data['Tifa']).fit(smoothing_level=alpha,optimized=False,use_brute=True).fittedvalues
data[['Tifa','HWES1']].plot(title='Holt Winters Single Exponential Smoothing')
```
